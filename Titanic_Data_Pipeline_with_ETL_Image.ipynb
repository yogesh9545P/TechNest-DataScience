```python
# ðŸ“Œ Step 1: Import Libraries
import pandas as pd
import numpy as np
```

```python
# ðŸ“Œ Step 2: Load CSV File
df = pd.read_csv("C:\\Users\\yoges\\Downloads\\Titanic-Csv.csv")  
df.head()
```

```python
# ðŸ“Œ Step 3: Clean Missing Data
df.drop(columns=["Cabin"], inplace=True)
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
df.isnull().sum()
```

```python
# ðŸ“Œ Step 4: Encode Categorical Variables
df = pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first=True)
df.head()
```

```python
# ðŸ“Œ Step 5: Engineer New Features
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
df.head()
```

```python
# ðŸ“Œ Step 6: Export Cleaned Dataset
df.to_csv("cleaned_titanic.csv", index=False)
print("âœ… Cleaned Titanic dataset saved as 'cleaned_titanic.csv'")
```
